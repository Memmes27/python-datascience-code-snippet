```python
import pandas as pd 

df = pd.DataFrame({'a': [1, 2, 3, 4], 'b': [2, 3, 4, 6]})
diff = df.diff()
print(diff)
```

```python
shift_down = diff.shift(-1)
shift_down
```

```python
shift_up = diff.shift(1)
shift_up
```

```python
processed_df = shift_up.dropna().reset_index(drop=True)
print(processed_df)
```
