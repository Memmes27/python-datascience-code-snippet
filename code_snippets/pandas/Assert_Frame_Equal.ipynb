```python
from pandas.testing import assert_frame_equal
import pandas as pd

# Importing the necessary libraries

df1 = pd.DataFrame({'coll': ['MO','Jas','Alondra'], 'col2': ['Yesi','Vannesa','Ximena']})
df2 = pd.DataFrame({'coll': ['MO','Jas','Alondra'], 'col2': ['Yesi','Vannesa','Ximena']})

# Creating two DataFrames, df1 and df2, with different values in the 'coll' column

assert_frame_equal(df1, df2)

# Asserting that the two DataFrames, df1 and df2, are equal. If they are not equal, an AssertionError will be raised.
```
